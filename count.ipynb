```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/home/khundmeer/Desktop/coins.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
plt.imshow(blur, cmap='gray')
```

```python
# canny = cv2.Canny(blur, 7, 7, 3)
ret,thresh=cv2.threshold(blur,240,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh, cmap='gray')
```

```python
# dilated = cv2.dilate(thresh, (1, 1), iterations=0)
erode=cv2.erode(thresh,(3,3),iterations=5)
plt.imshow(erode, cmap='gray')
```

```python
(cnt, hierarchy) = cv2.findContours(
	erode.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)

plt.imshow(rgb)
```
